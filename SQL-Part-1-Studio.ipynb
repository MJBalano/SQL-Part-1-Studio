## Part 1. BooksDB Questions

```sql
-- return top 100 results
SELECT TOP (100) book_id, authors AS "Author", title AS "Title", average_rating AS "Average Rating"
FROM BooksDB.dbo.books
ORDER BY average_rating DESC;
```

```sql
-- return least popular book
SELECT book_id, authors AS "Author", title AS "Title", average_rating AS "Average Rating"
FROM BooksDB.dbo.books
ORDER BY average_rating ASC;
```

Question 5: How many books where released in the first decade of 2000?

```sql
-- return number of books published in the first decade of 2000
SELECT COUNT(original_publication_year) AS "Number of Books Published in the First Decade of 2000"
FROM BooksDB.dbo.books
WHERE (original_publication_year) BETWEEN 2000 and 2010;
```

Question 6: How many book titles contain the word Happy?

```sql
-- return number of books that contain the word Happy
SELECT COUNT(*)
FROM BooksDB.dbo.books
WHERE title LIKE 'Happy%';
```

Question 9: Write a query that returns the number of Authors whose first name is between rock and roll.

```sql
-- return number of authors whose first name is between rock and roll
SELECT book_id, authors AS "Author", title AS "Title", average_rating AS "Average Rating"
FROM BooksDB.dbo.books
WHERE authors <= 'roll' AND authors >= 'rock'
ORDER BY authors ASC;
```

Your Question:  Write a query that returns the longest title?

```sql
-- Add your querry here
SELECT TOP (50) book_id, authors AS "Author", title AS "Title", average_rating AS "Average Rating",LEN(title) AS "Length of Title"
FROM BooksDB.dbo.books 
ORDER BY LEN(title) DESC;
```

```sql
--Removing spaces in title
SELECT TOP(100) book_id, authors AS "Author", title AS "Title", average_rating AS "Average Rating", LEN(REPLACE(title, ' ', '')) AS "Length of Title"
FROM BooksDB.dbo.books 
ORDER BY LEN(REPLACE(title, ' ', '')) DESC;
```

Your Question: Top  ten books written the year I was born?

```sql
-- Add your querry here
SELECT TOP 10 title AS "Title", authors AS "Author", original_publication_year AS "Publication Year", average_rating AS "Average Rating"
FROM BooksDB.dbo.books
WHERE original_publication_year = 1988
ORDER BY average_rating DESC;
```
